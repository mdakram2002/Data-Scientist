```python
import cv2
import numpy as np
```

```python
org_img = cv2.imread("img1.jpg")
rev_img = cv2.resize(org_img, (500, 500))
w,h = rev_img.shape[0],rev_img.shape[1]


m = cv2.getRotationMatrix2D((w/2, h/2), 90, 1)
new_img = cv2.warpAffine(rev_img, m, (w, h))
# new_img

h = np.hstack((rev_img, new_img))
cv2.imshow("show", h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
