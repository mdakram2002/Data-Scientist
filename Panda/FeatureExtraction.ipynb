```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if check == ")":
            check = False
            break
        if i == "(":
            check = True
        if check == True:
            data += i
    return data
```

```python
df["Title"].apply(extrac_episodes)
```
