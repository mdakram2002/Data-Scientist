```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Store': ['S1', 'S2', 'S1', 'S3', 'S2', 'S1', 'S3', 'S2'],
    'Sales': [120, 180, 160, 240, 210, 310, 270, 350],
    'Quantity': [12, 18, 15, 22, 20, 25, 30, 28],
    'Date': pd.date_range('2023-02-01', periods=8)
}

df = pd.DataFrame(data)
df
```

```python
#GROUP BY CATEGORY AND CALCULATE THE SUM OF SALES
cat = df.groupby('Category')
cat
for i, v in cat:
    print(i)
    print(v)
```

```python
#GROUP BY CATEGORY AND CALCULATE THE SUM OF SALES
cat = df.groupby('Category')['Sales'].sum()
cat
```

```python
#GROUP BY CATEGORY AND CALCULATE THE SUM OF SALES
cat = df.groupby('Store')['Sales'].sum()
cat
```
