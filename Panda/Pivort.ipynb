```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Date': pd.date_range('2023-03-01', periods=24),
    'Product': ['X', 'Y', 'Z', 'X', 'Y', 'Z'] * 4,
    'Region': [
        'North', 'South', 'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South', 'East', 'West',
        'North', 'South', 'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South', 'East', 'West'
    ],
    'Sales': np.random.randint(150, 1200, 24),
    'Units': np.random.randint(5, 60, 24),
    'Rep': [
        'Amit', 'Neha', 'Rohit', 'Pooja', 'Amit', 'Neha',
        'Rohit', 'Pooja', 'Amit', 'Neha', 'Rohit', 'Pooja',
        'Amit', 'Neha', 'Rohit', 'Pooja', 'Amit', 'Neha',
        'Rohit', 'Pooja', 'Amit', 'Neha', 'Rohit', 'Pooja'
    ]
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)

df
```

```python
pd.pivot_table(df, values='Sales', index='Region', columns='Product')
```

```python
pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='median')
```

```python
pivot2 = pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product')
pivot2
```

```python
#CROSS TABLE
pd.crosstab(df['Region'], df['Product']) # crosstav is used for counting purpose
```
