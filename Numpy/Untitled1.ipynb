```python
img = cv2.imread("img1.jpg")
img = cv2.resize(img, (500, 500))
print(img.shape)

new_img = cv2.Canny(img, 100, 200)
print(new_img.shape)

cv2.imshow("Show", new_img)
cv2.imshow("Show1", img)

cv2.waitKey(0)
cv2.destroyAllWindows()

(500, 500, 3)
(500, 500)
```
