```python
import cv2
import numpy as np
```

```python
# Median Blur
org_img = cv2.imread("img2.jpg")
rev_img = cv2.resize(org_img,(400, 400))

g = cv2.GaussianBlur(rev_img,  (7, 7), 0)
h = np.hstack((rev_img, g))

cv2.imshow("Show", h)
# cv2.imshow("Show1", org_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
